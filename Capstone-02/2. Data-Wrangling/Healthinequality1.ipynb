**Filling missing values in table 10 and table 12**

A column is removed if more than 20% of the values are missing

A state is removed if all the values of a column are missing

Missing values are filled with the mean value for that respective state

```python
# Choosing useful colums in table 10
ntable10_1 = table10.loc[:, 'cz':'stateabbrv']
ntable10_2 = table10[['cur_smoke_q1', 'bmi_obese_q1', 'exercise_any_q1']]
ntable10_3 = table10.loc[:, 'puninsured2010':'adjmortmeas_pnall30day']
ntable10_4 = table10.loc[:, 'cs00_seg_inc':'unemp_rate']
ntable10 = pd.concat([ntable10_1, ntable10_2, ntable10_3, ntable10_4], axis = 1)
```

```python
ntable10.head(10)
```

```python
# Removing the columns whose empty values are more than 20%
l = len(ntable10)
names = ntable10.columns.values[6:] 
for name in names:
    n = ntable10[name].isnull().values.sum()
    if n/l > 0.2: 
        ntable10 = ntable10.drop([name], axis=1)
```

```python
names = ntable10.columns.values[6:]
names
```

```python
states = ntable10['statename'].unique()
states
```

```python
for name in names:
    g = ntable10.groupby(['statename'])
    m = g[name].mean()
    states = ntable10['statename'].unique()
    for state in states:
        if m.isnull()[state]:
            ntable10 = ntable10[ntable10.statename != state]
        else:
            t = ntable10[ntable10.statename == state]
            t = t[name].fillna(m[state])
            ntable10.loc[ntable10.statename == state, name] = t
ntable10.to_csv('table_10.csv', index=False)
```
