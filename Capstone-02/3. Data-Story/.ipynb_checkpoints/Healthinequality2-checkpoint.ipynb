```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# counting number of male
count_m = table1[table1.gnd == 'M']['count'].sum()
print('Count of Male:', count_m)
```

```python
# counting number of female
count_f = table1[table1.gnd == 'F']['count'].sum()
print('Count of Female:', count_f)
```

```python
table1.groupby(by='gnd')['count'].sum().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Counts of male & female in pooling 2001~2014')
plt.show()
```

```python
# Counting of Household Income Percentile
table1.plot(kind='scatter', x='pctile', y='count')
plt.xlabel('Household Income Percentile')
plt.ylabel('Count')
plt.title('Count of Household Income Percentile')
plt.show()
```

```python
# distribution of Mean Household Income
table1.plot(kind='scatter', x='pctile', y='hh_inc', marker='.')
plt.xlabel('Household Income Percentile')
plt.ylabel('Mean Household Income')
plt.title('Distribution of Household Income by Percentile')
plt.show()
```

```python
# distribution of life expectancy of male
table1_m = table1[table1.gnd == 'M']
plt.hist(table1_m.le_raceadj, bins=25, color=sns.color_palette("Set2", 1))
plt.title('Life Expectancy of Male')
plt.show()
```

```python
# distribution of life expectancy of female
table1_f = table1[table1.gnd == 'F']
plt.hist(table1_f.le_raceadj, bins=25, color=sns.color_palette("Set1", 1))
plt.title('Life Expectancy of Female')
plt.show()
```
